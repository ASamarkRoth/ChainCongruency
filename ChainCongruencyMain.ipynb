# Schmidt tests on the set of short E115 decay chains

## The Schmidt test

Around the turn of the millenium [Schmidt et.al](https://link.springer.com/article/10.1007/s100500070129) suggested a new approach on how to assess decay chains. 

In the, so called _Schmidt test_, the logarithm of the lifetimes $\theta = \ln t$ and the expected value of its standard deviation ($\sigma_\theta$) of one decay step is considered. 
The fundamental principle of the test is that the shape of the probability density distribution and in continuation the expected value of the standard deviation is independent of the decay constant $\lambda$ of the radioactive species in study.

For an infinite number of events, i.e. an infinite number of measured life times, analytically the expected value of the standard deviation is; $\sigma_\theta = 1.28 \pm 2.15 \cdot \sqrt{m}$ where $m$ is the number of events.

The interpretation of the obtained standard deviation is as follows: 
* A too large value indicates that the set of life times do not originate from one single radioactive species.
* A too small value indicates a lack of sensitivity in the measurements or that the data is unjustly pruned.

If the _short_ chains of E115 is considered then we arrive at the following values for the Schmidt test: 

```python
Schmidt = np.nanstd(np.log(l_df_short[0]), axis=0)
col_names = []
[col_names.append("Step "+str(i+1)) for i in range(len(Schmidt))]
Schmidt = dict(zip(col_names, Schmidt))
df_Schmidt = pd.DataFrame(data=Schmidt, index=["Experimental Schmidt value"], columns=col_names)
df_Schmidt
```

Through Monte-Carlo simulations the distribution of this standard deviation could be obtained as a function of the number of events measured. 
[Schmidt et.al](https://link.springer.com/article/10.1007/s100500070129) (Table 1) presents expected values and 90% confidence intervals for $\sigma_\theta$ for a variety of number of events $m$. 

Let's add these to the table above. 

```python
Schmidt_conf = [[1.19, 1.19, 1.16], ["[0.73, 1.77]", "[0.73, 1.77]", "[0.65, 1.82]"]]
df_Schmidt.append(pd.DataFrame(data=Schmidt_conf, index=["Expected value of $\sigma_θ$", "90% confidence interval"], columns=col_names))
```

The Schmidt values from the experimental data does fit in the 90% confidence interval and thus it is justified to draw the conclusion that the data (for each decay step) originates from only one radioactive species. 

## Generalised Schmidt test

The Schmidt method is limited to one decay step at a time and hence not all data available is considered. If all steps of the decay chains have common origins a stricter test would be to group all steps of each chain together. An arithmetic average is not sufficient. 

A generalisation of the Schmidt test was proposed in [D. Rudolph et al., EPJ Web of Conferences](https://www.epj-conferences.org/articles/epjconf/pdf/2016/12/epjconf_nn2016_01001.pdf) and it is thoroughly described in [Ulrika Forsberg's PhD thesis](http://portal.research.lu.se/portal/files/7495513/thesis.pdf). 

Assume that $m$ chains have been observed and that each chain $i$ contains $n_i$ decay steps. Let $\theta_{i_j}$ be the logarithm of the $j$:th lifetime in chain $i$. Then the measure: 

\begin{equation}
\xi_{m, n} = \sqrt[2]{\frac{\sum\limits_{i=1}^m \sqrt[n_i]{\prod\limits_{j=1}^{n_i} \left( \theta_{i_j} - \bar{\theta_j} \right)^2  }  }{m}}, \bar{\theta_j} = \frac{ \sum\limits_{i=1}^m \theta_{i_j} } {m}
\end{equation}

does incorporate correlation times along decay chains instead of between single decay steps within a set of chains. 

## The generalised Schmidt method applied

This section deals with the implementation of the generalised Schmidt method. 

To calculate the geometrical mean in an array with `NaN` values the following (non-generic) function is defined.

```python
def g_nan_mean(data):
    if len(np.shape(data)) == 1:
        return data
    ret = np.empty(np.shape(data)[0])
    for i in range(np.shape(data)[0]):
        temp = 1
        steps = 0
        for j in range(np.shape(data)[1]):
            if isinstance(data, pd.DataFrame) and ~np.isnan(data.iloc[i,j]):
                temp *= data.iloc[i,j]
            elif not isinstance(data, pd.DataFrame) and ~np.isnan(data[i,j]):
                temp *= data[i,j]
            else:
                break
            steps += 1
        ret[i] = temp**(1./steps)
    return ret
```

The generalised Schmidt value is calculated for the two sets of decay chains; 
* Complete E115 Short chains.
* E115 Short chains with D3 excluded.

```python
gen_Schmidt = []
for dfs in l_df_short:
    #display(dfs)
    theta = np.log(dfs)
    theta_var = np.square(theta - np.nanmean(theta, axis=0))
    gen_Schmidt_temp = g_nan_mean(theta_var)
    gen_Schmidt.append(np.sqrt(np.mean(gen_Schmidt_temp)))

col_names = ["E115 Short chains", "Dubna 3 Excl."]
df_gen_Schmidt = dict(zip(col_names, gen_Schmidt))
df_gen_Schmidt = pd.DataFrame(data=df_gen_Schmidt, index=["Experimental Generalised Schmidt value"], columns=col_names)
df_gen_Schmidt
```

### Simulation of expected values and confidence intervals

Via Monte-Carlo simulations the expected value and its confidence intervals can be obtained for the generalised Schmidt method. 10 000 sets of decay chains with the same structure as the experimental sets extracted above. Each decay step is simulated from an exponential distribution with an arbitrary decay constant (here $\lambda = 1$). Following this the exact same calculations are made as above. To get an idea of the behaviour of the method, a different number of decay steps, $n$, are included. 4 out of the 14 short chains only have two decay steps and to see the effect another index $l$ is introduced which represents the number of chains which have only two decay steps.  

```python
sim_names = ["$m=14$, $n=1$", "$m=14$, $n=2$", "$m=14$, $n=3$", "$m=14$, $n=1$, $l=4$", "Excl. D3"]
nbr_sets = 10000
df_temp = np.zeros(np.shape(l_df_short[0]))
l_df_sim = [df_temp[:,0], df_temp[:,0:2], df_temp, l_df_short[0], l_df_short[1]]

sim_Schmidt = np.empty((nbr_sets, len(sim_names)))

for i, df_sim in enumerate(l_df_sim):
    print("Set: ", i, sim_names[i])
    shape = (nbr_sets, *np.shape(df_sim))
    sim = np.random.exponential(scale=1, size=shape)
    if np.count_nonzero(np.isnan(df_sim)) > 0:
        for j in range(nbr_sets):
            sim[j][np.isnan(df_sim)] = np.nan
    theta = np.log(sim)
    theta_var = np.zeros(np.shape(theta))
    temp_Schmidt = np.zeros(np.shape(theta)[0:2])
    for j in range(nbr_sets):
        theta_var[j] = np.square(theta[j] - np.nanmean(theta[j], axis=0))
        temp_Schmidt[j] = g_nan_mean(theta_var[j])
    sim_Schmidt[:, i] = np.sqrt(np.mean(temp_Schmidt, axis=1))

sim_final_Schmidt = np.mean(sim_Schmidt, axis=0)
print(sim_final_Schmidt)
print("Done")
```

What follows is an illustration of the effect of the generalised Schmidt method (cf. Figure 3. in [D. Rudolph et al., EPJ Web of Conferences](https://www.epj-conferences.org/articles/epjconf/pdf/2016/12/epjconf_nn2016_01001.pdf)).

```python
plt.figure(figsize=(12,10))
for i, l in enumerate(sim_names[0:-1]):
    plt.hist(sim_Schmidt[:,i], bins=100, color=colors[i], label=l, histtype='step', normed=True, lw=2)
plt.legend(loc='best',markerfirst=True, frameon=True,shadow=True,fancybox=True)
plt.xlabel(r"$\xi_{m,n,l}$")
plt.xlim((0.3,2.5))
plt.ylabel("Intensity")
plt.title(r"Distributions of $\xi_{m,n,l}$ for different sets of decay chains")
plt.tight_layout()
plt.show()
```

The more steps that are included (larger $n$) the smaller the distribution of $\xi_{m,n,l}$ becomes. Note that the blue curve corresponds to the standard Schmidt test.

# Figure-of-Merit (FoM) method 

Described in [U. Forsberg et al., Nucl. Phys. A](http://www.sciencedirect.com/science/article/pii/S0375947416300768?via%3Dihub):
> A FoM$^{(n)}_j$, defined for each correlation time $t^{(n)}_j$ decay step $j = 1, 2, 3$ of the chain identified by the number $n$, is calculated as the value of a probability density function for a reference data  set. The geometric mean of FoM$^{(n)}_j$ over all available steps $j$ in chain $n$ defines the FoM$^{(n)}_{geom}$ for that chain. The arithmetic mean of FoM$^{(n)}_{geom}$over all $N$ chains defines the FoM for the data set with respect to the interpretation under consideration. 

> If the individual chains all deviate strongly from the average data, the FoM value will be low. If the chains are all too similar to their average behaviour, the FoM will be high. The test is similar to the Schmidt test described above.  Note, however, that a low $\sigma_\theta$ corresponds to a large FoM and vice versa.


> The probability density function is obtained through the following: 

> 1. For each step $j$, the average experimental lifetime $\bar{t}_j$ calculated, and the number of available lifetimes $N_j$ is noted. 
> 2. For each step $j$, the likelihood function for the true lifetime $\tau_j$, given by $N_j$ and $\bar{t}_j$, is  determined. 
> 3. For each step $j$, a $\tau_j$ is selected with a probability governed by the likelihood function for $\tau_j$, and then a set of $N_j$ lifetimes are generated from the exponential distribution defined by this $\tau_j$. This procedure is repeated until a smooth histogram emerges.

The analytical expression for this _smeared_ PDF for step $j$, using a reference data set with $N_j$ data points and average lifetime $\bar{t}_j$ in step $j$: 

\begin{equation}
f(t)=t(N_j−1) \frac{ (N_j \bar{t}_j)^{N_j-1} }{ (N_j \bar{t}_j + t)^{N_j} }
\end{equation}

and it is obtained by weigthing an exponential distribution $g(t) = \frac{t}{\tau} e^{- \frac{t}{\tau} }$ with the normalised likelihood function for $\tau$:

\begin{equation}
h(\tau)= \frac{ N^{N-1} }{ (N-2)! }\frac{ \bar{t}^{N-1} }{ \tau^N } e^{- \frac{N\bar{t}}{\tau}}
\end{equation}


The smeared probability density function eq. (2) is implemented as follows.

```python
def pdf_smeared(times):
    """ Smeared probability density function, eq. (2)"""
    if isinstance(times, pd.DataFrame):
        times = times.as_matrix()
    N_j = np.count_nonzero(~np.isnan(times), axis=0)
    t_bar = np.nanmean(times, axis=0)
    ret = np.zeros((len(times[:]), len(N_j)))
    for i, t in enumerate(times[:]):
        temp = np.empty(len(t))
        prod = np.empty(len(t))
        np.multiply(N_j, t_bar, out=prod)
        np.power(prod, np.subtract(N_j,1), out=temp) #numerator
        np.divide(temp, np.power(np.add(prod, t), N_j), out=temp) #division with denominator
        np.multiply(temp, np.multiply(t, np.subtract(N_j, 1)), out=temp)
        ret[i] = temp
    return ret
```

The FoM$_j$ and FoM$_{geom}$ are calculated for the two sets of decay chains. Simultaneously the results are written to a table (cf. Table 4. in [U. Forsberg et al., Nucl. Phys. A](http://www.sciencedirect.com/science/article/pii/S0375947416300768?via%3Dihub)).

```python
calc_names = ["E115 Short chains", "D3 Excl. "]
for i, df_now in enumerate(l_df_short):
    # Calculating FoM_j, FoM_geom and a final FoM for the complete set
    fom_j = pdf_smeared(df_now)
    fom_geom = g_nan_mean(fom_j)
    fom_final = np.mean(fom_geom)

    # Creating table in pandas data frame
    s_cols = 'FoM$_{}$'
    col_names = [s_cols.format(j) for j in range(len(fom_j[0]))]
    col_names.append('FoM$_{geom}$')
    header = pd.MultiIndex.from_product([[calc_names[i]], col_names])
    data = dict(zip(header, np.append(fom_j, fom_geom[:,np.newaxis], axis=1).T))
    df_temp = pd.DataFrame(data=data, index=df_now.index, columns=header)
    data = dict(zip(header, [ None, None, None, fom_final]))
    df_temp = df_temp.append(pd.DataFrame(data=data, index=[("FoM", '')], columns=header))
    if i == 0:
        df_fom = df_temp
    else:
        df_fom = pd.concat([df_fom, df_temp], axis=1, join_axes=[df_fom.index])
df_fom = df_fom.round(decimals=3)
display(df_fom)
```

## Simulations

### Simulating sets of decay chains

The sets of decay chains are simulated in follows.

For each decay step that exist in the set of decay chains:
1. For the decay step considered, a random $\tau_1$ is picked according to the $\tau$ likelihood function $h(\tau)$ (see eq. (3)). 
2. A number of random lifetimes (corresponding to the number of lifetimes that exist in the studied set of decay chains) from the exponential distribution $g(t)$ defined by this $\tau_1$ is generated. 

The $\tau$ likelihood function is first defined. In the following, all steps are treated simultaneously. 

```python
from scipy.special import factorial
class tau_likelihood_func(object):
    def __init__(self, times):
        if isinstance(times, pd.DataFrame):
            times = times.as_matrix()
        self.N_j = np.count_nonzero(~np.isnan(times), axis=0)
        self.t_bar = np.nanmean(times, axis=0)
        self.factor1 = np.divide(np.multiply(np.power(self.N_j, self.N_j-1), np.power(self.t_bar, self.N_j-1)), factorial(self.N_j-2))
        self.factor2 = -np.multiply(self.N_j, self.t_bar)
        #print("factor1=", self.factor1, np.shape(self.factor1))
                                 
    def __call__(self, tau):
        #ret = np.empty((len(tau), len(self.N_j)))
        #print(tau, np.shape(tau))
        ret = np.multiply(np.divide(self.factor1, np.power(tau, self.N_j)), np.exp(np.divide(self.factor2, tau)))
        return ret

# Two different tau likelihood functions are defined corresponding to the complete set of E115 short chains and the set with D3 excluded
tau_likelihoods = [tau_likelihood_func(l_df_short[0]), tau_likelihood_func(l_df_short[1])]
```

#### 1. Random sampling of $\tau$ from its likelihood function

A numerical cumulative density function (CDF) is created from which the random samples of $\tau$ are generated. 

```python
# Creating points for the numerical CDFs
nbr_points = 10000
taus = []
tau_ls = []
cdfs = []

# looping sets of decay chains
for i, df_now in enumerate(l_df_short):
    print("t_mean=", tau_likelihoods[i].t_bar)
    tau = np.empty((nbr_points, len(df_now.columns)))
    # looping number of steps in set
    for j in range(len(tau[0])):
        tau[:,j] = np.linspace(0.00001, tau_likelihoods[i].t_bar[j]*4, len(tau[:,0]))
    tau_l = tau_likelihoods[i](tau)

    # Creating normalised CDFs 
    cdf = np.cumsum(tau_l, axis=0)
    cdf = np.divide(cdf, cdf[-1,:])
    
    # Storing in lists
    taus.append(tau)
    tau_ls.append(tau_l)
    cdfs.append(cdf)
    
# Plotting the CDF
labels = ["Step "+str(i+1) for i in range(len(l_df_short[0].columns))]
plt.figure()
for i, l in enumerate(labels):
    plt.plot(taus[0][:,i], cdfs[0][:,i], label=l)
plt.legend(loc='best',markerfirst=True, frameon=True,shadow=True,fancybox=True)
plt.xlabel(r"$\tau$ (s)")
plt.ylabel(r"CDF")
plt.title("Numerical CDF for the 3 decay steps")
plt.tight_layout()
plt.show()
```

Random sampling of $\tau$ from the numerical CDF for 100 000 sets of decay chains. 

```python
nbr_sets = 100000

rand_taus = []
# looping sets of decay chains
for ind, df_now in enumerate(l_df_short):
    #display(df_now)
    # Generating random numbers from uniform [0, 1)
    rands = np.random.rand(nbr_sets, len(df_now.columns))

    # Get the corresponding taus from the CDF
    rand_tau = np.empty(np.shape(rands))
    #looping number of steps in set
    for i in range(len(df_now.columns)):
        inds = np.searchsorted(cdfs[ind][:,i], rands[:,i])
        rand_tau[:,i] = taus[ind][inds,i]
    rand_taus.append(rand_tau)
```

Plotting the sampled $\tau$ for the existing decay steps of the complete set of E115 short chains.

```python
f, axes = plt.subplots(3, 1, figsize=(10,8))# sharex=True, sharey=True)

# add a big axes, hide frame
f.add_subplot(111, frameon=False)
# hide tick and tick label of the big axes
plt.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')

# Plotting the sampled taus and their pdf:s
for i,ax in enumerate(axes):
    ax.plot(taus[0][:,i],tau_ls[0][:,i], color='black', label="Real PDF")
    ax.hist(rand_taus[0][:,i], bins=100, color=colors[i], normed=True, label="Sampled Distribution")
    #ax.annotate(an[i],xy=(0.75,.8), fontsize=24, xycoords='axes fraction', color='k')
    ax.legend(loc='upper right',markerfirst=False, title=r"$\bf{"+labels[i]+"}$", frameon=True, shadow=True, fancybox=True, fontsize=16)
    #ax.set_xscale('log')
    #ax.set_xlim(0.2)
plt.ylabel("PDF")
plt.xlabel(r"$\tau$ (s)")
plt.tight_layout()
plt.show()
```

#### 2. Simulating 100 000 sets of decay chains AND calculating their FoM values.
This could take a minute or two.

```python
sim_names = ["E115 Short Chains", "D3 Excl."]

l_df_sim = [l_df_short[0], l_df_short[1]]

sim_FoM = np.empty((nbr_sets, len(sim_names)))

#Looping over data sets as defined in l_df_sim
for i, df_sim in enumerate(l_df_sim):
    print("Set: ", i, sim_names[i])
    
    # Simulations of all random generated sets of decay chains
    shape = (nbr_sets, *np.shape(df_sim))
    print("Shape of current set of decay chains: ", np.shape(df_sim))
    sim = np.empty(shape)
    
    #looping over sets of decay chains
    for j in range(nbr_sets)):
        #looping over steps in current set of decay chains and adding simulated lifetimes
        for k in range(len(df_sim.columns)):
            sim[j, :, k] = np.random.exponential(scale=rand_taus[i][j, k], size=shape[1])
    # if the current set of decay chains contains NaN (e.g. chains with one less decay step), substitute these in the simulated data
    if np.count_nonzero(np.isnan(df_sim)) > 0:
        for j in range(nbr_sets):
            sim[j][np.isnan(df_sim)] = np.nan
    print("    Simulated all sets of decay chains")
    print("     sim[0:2]=\n", sim[0:2])
    # FoM calculations
    fom = np.empty(shape[1:])
    fom_geom = np.empty(shape[1])
    for j in range(nbr_sets):
        fom = pdf_smeared(sim[j])
        fom_geom = g_nan_mean(fom)
        sim_FoM[j,i] = np.mean(fom_geom)
print("Done")
sim_FoM_mean = np.mean(sim_FoM, axis=0)
```

# The alleged link between Element 117 and 115

The fourth IUPAC/IUPAP Joint Working Party published a report with a review of the claims of the discovery of the new elements 113, 115, 117 & 118 [[Report](https://www.degruyter.com/view/j/pac.2016.88.issue-1-2/pac-2015-0502/pac-2015-0502.xml)]. In the report the joint working party concludes that the collaborations claims are in compliance with the set criteria. 

As stated in [U. Forsberg et al., Physics Letters B, ](http://www.sciencedirect.com/science/article/pii/S0370269316303495?via%3Dihub): 

> One indirect method to support that a new element was produced is via cross reactions or cross bombardments. The idea is that the same $\alpha$-decay chain is entered by two or more different nuclear reactions. For example, the $\alpha$-decay chains comprising the superheavy nuclei $^{289}$115, $^{285}$113 and $^{281}$Rg are considered to have been populated following two different fusion-evaporation reactions, namely via $^{243}$Am($^{48}$Ca,2n)$^{289}$115 and $^{249}$Bk($^{48}$Ca,4n)$^{293}$117. 

The two main properties of a decay chain are the measured $\alpha$ energies and lifetimes of the radioactive isotopes. 
In the following it is evaluated of whether:
> The cross reaction case for the isotopes $^{293}$117, $^{289}$117, $^{285}$113 and $^{281}$Rg presented in the IUPAC/IUPAP Joint Working Party [[Report](https://www.degruyter.com/view/j/pac.2016.88.issue-1-2/pac-2015-0502/pac-2015-0502.xml)] is probable.

I.e. we want to evaluate of whether decay chains assigned to the isotopes $^{293}$117, $^{289}$117, $^{285}$113 and $^{281}$Rg have a common origin. 
Is there a link between Element 117 and 115?

The assessment will be performed with the statistical FoM method presented above.
Hence, lifetimes are merely considered.

However, is such an interpretation backed up by the statistical methods walked through in this document?

We create four different sets of data: 
1. All E117 chains (denoted E117).
2. All Dubna Short E115 and all E117 chains (denoted D & E117).
3. All Dubna Short E115 (D3 excluded) and all E117 chains (denoted D' & E117).

```python
names_df = ["E117", "D & E117", "D' & E117", "D3 & E117"]
l_df_links = [df_links_all.loc['E117'], df_links_all, df_links_all.drop(index=[('E115', '3')]), df_links_all.drop(index=[('E115', '1'), ('E115', '2'), ('E115', '4')])]
#display(df_links[3])
```

The FoM method is used and for every set:
1. The total FoM for is calculated as in section 4. 
2. Confidence intervals are calculated by the means of Monte-Carlo simulations as described in section 4.1.

## 1. Calculations

```python
# Step 1; Calculations
fom_calc = np.empty(len(names_df))
for i, df_now in enumerate(l_df_links):
    fom_j = pdf_smeared(df_now)
    fom_geom = g_nan_mean(fom_j)
    fom_calc[i] = np.mean(fom_geom)
links_calc = pd.DataFrame(data={"FoM": fom_calc}, index=names_df)
links_calc.index.rename('Data set', inplace=True)
print(links_calc.columns)
display(links_calc.round(decimals=3))
```

## 2. Simulations

```python
# Step 2; Simulations

tau_likelihoods = [] 
for df_now in l_df_links:
    tau_likelihoods.append(tau_likelihood_func(df_now))

# Creating points for the numerical CDFs
nbr_points = 10000
taus = []
tau_ls = []
cdfs = []
for i, df_now in enumerate(l_df_links):
    tau = np.empty((nbr_points, len(df_now.columns)))
    for j in range(len(tau[0])):
        tau[:,j] = np.linspace(0.00001, tau_likelihoods[i].t_bar[j]*4, len(tau[:,0]))
    tau_l = tau_likelihoods[i](tau)

    # Creating normalised CDFs 
    cdf = np.cumsum(tau_l, axis=0)
    cdf = np.divide(cdf, cdf[-1,:])
    
    # Storing in lists
    taus.append(tau)
    tau_ls.append(tau_l)
    cdfs.append(cdf)
    
print("Done")
```

Random sampling from the numerical CDF for 100 000 sets of decay chains. 

```python
nbr_sets = 100000

# Generating random numbers from uniform [0, 1)
rand_taus = []
for ind, df_now in enumerate(l_df_links):
    rands = np.random.rand(nbr_sets, len(df_now.columns))

    # Get the corresponding taus from the CDF
    rand_tau = np.empty(np.shape(rands))
    for i in range(len(df_now.columns)):
        inds = np.searchsorted(cdfs[ind][:,i], rands[:,i])
        rand_tau[:,i] = taus[ind][inds,i]
    rand_taus.append(rand_tau)

print("Done")
```

Simulating 100 000 sets of decay chains and calculating their FoM values. This could take a few minutes.

```python
sim_FoM = np.empty((nbr_sets, len(names_df)))

#Looping over data sets as defined in l_df_sim
for i, df_sim in enumerate(l_df_links):
    print("Set: ", i, names_df[i])
    
    # Simulations of all random generated sets of decay chains
    shape = (nbr_sets, *np.shape(df_sim))
    sim = np.empty(shape)
    
    #looping over sets of decay chains
    for j in range(len(sim_FoM[:,0])):
        #looping over steps in current set of decay chains and adding simulated lifetimes
        for k in range(len(df_sim.columns)):
            sim[j, :, k] = np.random.exponential(scale=rand_taus[i][j, k], size=shape[1])
    # if the current set of decay chains contains NaN (e.g. chains with one less decay step), substitute these in the simulated data
    if np.count_nonzero(np.isnan(df_sim)) > 0:
        for j in range(nbr_sets):
            sim[j][np.isnan(df_sim)] = np.nan
    print("    Simulated all sets of decay chains")
    
    # FoM calculations
    fom = np.empty(shape[1:])
    fom_geom = np.empty(shape[1])
    for j in range(nbr_sets):
        fom = pdf_smeared(sim[j])
        fom_geom = g_nan_mean(fom)
        sim_FoM[j,i] = np.mean(fom_geom)
sim_FoM_mean = np.mean(sim_FoM, axis=0)
print("Done")
```
